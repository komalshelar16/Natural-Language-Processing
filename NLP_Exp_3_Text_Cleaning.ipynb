```python
data_in['char_counts'] = data_in['twitts'].apply(lambda x:len(re.sub(pattern,'',x)))
data_in
```

```python
data_in['word_counts'] = data_in['twitts'].apply(lambda x:len(x.split()))
data_in.sample()
```

```python
data_in['avg_word_length'] = data_in['char_counts']/data_in['word_counts']
data_in.sample(3)
```

```python
data_in['avg_word_length'].apply(lambda x: round(x,2))
```

```python
data_in['hashtag_tags_count'] =data_in['twitts'].apply(lambda x:len(re.findall(r'#\w+',x)))
data_in.sample(3)
```

```python
data_in['mentions_count'] =data_in['twitts'].apply(lambda x:len(re.findall(r'@\w+',x)))
data_in
```
