```python
import pandas as pd
import numpy as np
import re
```

For removing email address

```python
pattern = r'\w+@\w+.\w+'
data_in['twitts'].apply(lambda x: re.findall(pattern=pattern,string=x))
```

```python
pattern = r'\b[A-Za-z0-9.+_-]+@[A-Za-z.-]+.[A-Za-z]{2,}\b'
data_in['twitts'].apply(lambda x: re.findall(pattern=pattern,string=x))
```

For removing any website links

```python
pattern = r'http\S+ |www\.\S+'
data_in['twitts'].apply(lambda x: re.findall(pattern=pattern,string=x))
```

```python
import re
x = 'www.google.com is the website for finding the sentence.'
pattern = r'www\.+\S+'
output = re.findall(pattern,x)
print(output)
```

For HTML

```python
pattern = r'\b<\S+\b'
data_in['twitts'].apply(lambda x: re.findall(pattern=pattern,string=x))
```

For numbers

```python
pattern = r'\b[0-9]+\b'
data_in['twitts'].apply(lambda x: re.findall(pattern=pattern,string=x))
```
