```python
import spacy
spacy.load("en_core_web_sm")
import en_core_web_sm
nlp = en_core_web_sm.load()
```

```python
doc = nlp("Hello, I am Komal. I am taking this NLP Course.")
print(doc)
for word in doc:
    print(word.text,word.is_alpha)
```

```python
print(doc)
for sent in doc.sents:
  print(sent)
```
